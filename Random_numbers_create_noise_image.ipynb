##Creating noise image using pytorch

```python
import matplotlib.pyplot as plt

#h,w of image 256*256

uniform_noise = torch.rand((256,256))
plt.imshow(uniform_noise,cmap='grey')
##grey scaleimage - one channel
plt.title('uniform noise image')
plt.show()
```

```python
import matplotlib.pyplot as plt

#h,w of image 256*256

uniform_noise = torch.rand((256,256))
plt.imshow(uniform_noise.numpy(),cmap='grey')
##grey scaleimage - one channel
plt.title('uniform noise image')
plt.show()
```

```python
import matplotlib.pyplot as plt

#h,w,c of image 256*256

uniform_noise = torch.rand((256,256,3))
plt.imshow(uniform_noise.numpy())
##grey scaleimage - one channel
plt.title('uniform noise image')
plt.axis('off')
plt.show()
```

```python
##properties

print('shape',uniform_noise.shape)

print('shape',uniform_noise.dtype)


print('shape',uniform_noise.ndim)

print('shape',uniform_noise.ndimension)

print('min value',torch.min(uniform_noise))

print('min value',torch.max(uniform_noise))
```
