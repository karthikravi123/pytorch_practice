```python
import torch
```

```python
tensor_1 = torch.tensor([1,2,3])
tensor_2 = torch.tensor([4,5,6])
print(tensor_1.ndim)
print(tensor_1.size())
```

```python
stacked_tenspr_dim_0 = torch.stack((tensor_1,tensor_2),dim=0)
print(stacked_tenspr_dim_0.shape)
print(stacked_tenspr_dim_0)
```

```python
stacked_tenspr_dim_0 = torch.stack((tensor_1,tensor_2),dim=1)
print(stacked_tenspr_dim_0.shape)
print(stacked_tenspr_dim_0)
```

```python
tensor_3 = torch.tensor([7,8,9])
stacked_tenspr_dim_0 = torch.stack((tensor_1,tensor_2,tensor_3),dim=1)
stacked_tenspr_dim_0
```

```python
##stacking 2D tensors

tensor1 = torch.tensor([[1,2],[3,4]])
tensor2 = torch.tensor([[5,6],[7,8]])

stack1 = torch.stack((tensor1,tensor2),dim=0)
stack1
```

```python
stack1 = torch.stack((tensor1,tensor2),dim=1)
stack1
```

```python
stack1 = torch.stack((tensor1,tensor2),dim=2)
stack1
```

```python
torch.cat((tensor1,tensor2),dim=0)
```
