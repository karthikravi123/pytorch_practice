```python
import torch
```

##data types
- floating - for moost deep learning tasks --> float32, float64,float16
(memory consumption is less than 32,64 --it will effect performance)
-- lower the precision faster the speed on cpu-computation
- integers - for categorical data and indices --> int32 ,int64,int8
- boolean - mask or logical operations --> True/False
- complex number - advanced computation -->complex64,complex126=8
--> numerical precision : float64 is more precise than float32

```python
default_tensor = torch.tensor([1.5,2.5,3.5])
##default tensor datatype is float32
```

```python
float_tensor = torch.tensor([1.5,2.4,3.4],dtype = torch.float64)
float_tensor.dtype
```

```python
float_tensor = torch.tensor([1.5,2.4,3.4],dtype = torch.float32)
float_tensor.dtype
```

```python
int_tensor = torch.tensor([1,2,3])
int_tensor.dtype
```

```python
torch.tensor([True,False,True],dtype=torch.bool)
```

```python
##convert

float_tensor = torch.tensor([0,1,3,0,-1])
int_tensor = float_tensor.to(torch.int64)
bool_tensor = int_tensor.to(torch.bool)
bool_tensor
```

Impact of datatype on memory


```python
float_32_tensor = torch.ones(1000,dtype=torch.float32)

float64_tensor = torch.ones(1000,dtype=torch.float64)

##tensor.element size ()
#element_size() --> give you size of one element in bytes
#n_element() --> gives the total number of elements in tensor
print("memory used by float32 tensor :",float_32_tensor.element_size()* float_32_tensor.nelement())
print("\n")
print("Memory used by float64",float64_tensor.element_size()* float64_tensor.nelement())
```

float32 - >default for most NN models

float64 --> high precision computation

int32 --> general purpose integer

int64 --> tensor indices
