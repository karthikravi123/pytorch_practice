#joining tensor

```python
##torch.cat()--> merges tensors along and existing dimension

tensor1 =torch.tensor([[1,2],[3,4]])
tensor2 = torch.tensor([[4,5],[6,7]])
contcat_tensor_row =torch.cat((tensor1,tensor2),dim=0)
contcat_tensor_col =torch.cat((tensor1,tensor2),dim=1)
print(contcat_tensor_row,"\n",contcat_tensor_col)
```

#stack -->created a new_dimension ,increases the tensor rank

```python
##torch.cat()--> merges tensors along and existing dimension

tensor1 =torch.tensor([[1,2],[3,4]])
tensor2 = torch.tensor([[4,5],[6,7]])
stack_tensor_row =torch.stack((tensor1,tensor2),dim=0)
stack_tensor_col =torch.stack((tensor1,tensor2),dim=1)
print(stack_tensor_row,"\n",stack_tensor_col)
```

#splitting

- torch.chunk() -->divides tensor into equal size chunks

- torch.split() --> allows uneven splitting based on size

```python
original_tensor = torch.arange(12)

chunks = torch.chunk(original_tensor,3,dim=0)
print(chunks)
```

```python
original_tensor = torch.arange(13)

chunks = torch.chunk(original_tensor,5,dim=0)
print(chunks)
```

```python
for chunk  in chunks:
  print(chunk)
```

```python
original_tensor = torch.arange(13)

split = torch.split(original_tensor,5,dim=0)
print(split)
```

```python
for splits  in split:
  print(splits)
```

Transposing and premuting

--transpose() -->swaps two dimension.
m x n --> n x m

-- permute() - rearranges all dimenstion in the specified order

```python
tensor_original = torch.arange(24).reshape(12,2)
print(tensor_original)
tensor_original.transpose(0,1)
```

```python
tensor_original = torch.arange(24).reshape(2,3,4)
print(tensor_original)
#[0,1,2]
permute_tensor = tensor_original.permute(2,0,1)
```
