```python
import torch
```

```python
zeros_tensor = torch.zeros(5)
```

```python
zeros_tensor
```

```python
zeros_tensors = torch.zeros(256,256)
print("multidimenstion",zeros_tensors,"\n","shape",zeros_tensor.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
##mention min and max orelse matplotlib by default will normalize o to 1
plt.imshow(zeros_tensors.numpy(),cmap='grey',vmin=0,vmax=1)
plt.title("dark image")
plt.axis('off')
plt.show()
```
