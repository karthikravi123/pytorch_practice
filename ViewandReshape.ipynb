view() and reshape() are used to change the shape of tensors,
The core idea is to transform a tensor into a new shape while keeping the data consistent

View - contigous data in memory

reshape - more flexibility

Contiguous
orignal: [[1,2,3],[4,5,6]] = 2 x 3
in-memory :  [1,2,3,4,5]

Non-contigous

original data: [[1,2,3],[4,5,6]]

transposed data : [[1,4],[2,5],[3,6]]

IN-memory: [1,2,3,4,5,6]

The data remains the same in memory but the access pattern changes


low performance comapred to contigous data

Reshape - non contigous data

it creates copy of data of contingous data

Reshape works with both contigous and non contigous data

```python
import torch
```

```python
tensor= torch.arange(12)
```

```python
reshaped_tensor_view = tensor.view(2,6)
reshaped_tensor_reshape = tensor.reshape(2,6)
```
