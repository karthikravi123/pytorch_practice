pytorch -Indexing Tensors
This will help us efficiently access ,manipulate and analyze data stored intensors



```python
import torch
```

```python
##accessing elements in a tensor

'''
tensor[row,column]
'''

tensor = torch.tensor([[10,20,30],[40,50,60],[70,80,90]])
print(tensor.ndim)
```

```python
print(tensor.shape)
```

Indexing

```python
print("elements at (0,1)",tensor[0,1])
```

```python
tensor[0]
```

```python
tensor[:,0]
```

Slicing

```python
print("first two rows:",tensor[:2])
```

```python
print("first two colums",tensor[:,:2])
```

```python
print(tensor)
```

```python
print(tensor[1,1:3])
```

```python
##(0,1)-20 (2,2)-90
tensor[[0,2],[1,2]]
```

```python
##boolean indexing

mask = tensor > 50
mask
```

```python
print(tensor[mask])
```

```python
##changing tensor values via indexing

print(tensor)
tensor[0,1] =25
print(tensor)
```

```python
tensor[:,0] = torch.tensor([100,200,300])
```

```python
tensor
```

Advance slicing with source dot index select

```python
indices = torch.tensor([0,2])
selected_rows = torch.index_select(tensor,dim=0,index=indices)
selected_rows
```

```python
indices = torch.tensor([0,2])
selected_rows = torch.index_select(tensor,dim=1,index=indices)
selected_rows
```

```python
##using tensor with steps

print(tensor[:,::2])
```

```python
##using tensor with steps

#print(tensor[::-1])
torch.flip(tensor,dims=(0,))
```
